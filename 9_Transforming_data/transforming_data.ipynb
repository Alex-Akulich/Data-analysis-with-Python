## Заполнение пропусков средними значениями

```python
df_4.isnull().sum() # Функция isnull проверяет пропуски в данных
```

```python
df_4.median() # Рассчитаем средние медианные значения по всем фичам, чтобы заменить ими пропуски
```

```python
import math
median_test_score = math.floor(df_4['Alcohol'].mean())
median_test_score
```

```python
df_4['Alcohol'] = df_4['Alcohol'].fillna(median_test_score)
df_4
```

```python
df_4.Unemployed.median() # Другой способ расчета медианы
```

```python
df_4['Unemployed'] = df_4['Unemployed'].fillna(df_4.Unemployed.median())
df_4
```
