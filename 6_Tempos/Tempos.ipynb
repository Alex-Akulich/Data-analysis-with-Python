```python
import pandas as pd
from pandas import DataFrame
from datetime import datetime
from datetime import date
```

```python
df = pd.read_csv('Data/avocado.csv', sep=',')
df
```

```python
df.info() # Характеристика данных (пропуски и тип). Видно, что колонка Date трактуется как Объект, т.е. текст, а не как дата
```

```python
type(df.Date[1]) # Проверяется тип данных в колонке Date. Видно, что эти данные воспринимаются как текст
```

## Готовится Data frame для расчетов темпа роста

```python
df2=df[['Date','AveragePrice']] # Создается новый Data frame
df2
```

### Пример 1. Расчет ежегодных темпов роста

```python
# Тип данных в колонке Date меняется с текстового на тип "дата и время"
df3=df2.copy()
df3['Date']=pd.to_datetime(df3['Date'])
df3
```

```python
type(df3.Date[0]) # Проверяется - поменялся ли тип данных в колонке Date?
```

```python
df3.info() # Еще один способ проверки: поменялся ли тип данных в колонке Date?
```

```python
df3['Year']=df3['Date'].dt.year # Добавляется колонка Year
df3.head()
```

```python
# Происходит группировка данных по годам и рассчитываются средние арифметические значения
df3=df3.groupby('Year')['AveragePrice'].mean()
df3
```

```python
df3=df3.reset_index() # Добавляется индекс (см. первую колонку)
df3
```

```python
df3['Difference']=df3['AveragePrice'].diff(1) # Рассчитываются абсолютные отклонения (колонка difference)
df3
```

```python
df3['YoY, in_percents']=df3['Difference']/df3['AveragePrice']*100+100 # рассчитываются темпы роста
df3
```

```python
# Округление до 2-х знаков после разделителя (см. колонку YoY,in_percents)
df3['YoY, in_percents']=round(df3['Difference']/df3['AveragePrice']*100+100,2)
df3
```

### Скрипт по расчету ежегодных темпов роста можно было разместить в одной ячейке

```python
df4=df2.copy()
df4['Date']=pd.to_datetime(df4['Date'])
df4['Year']=df4['Date'].dt.year
df4=df4.groupby('Year')['AveragePrice'].mean()
df4=df4.reset_index()
df4['Difference']=df4['AveragePrice'].diff(1)
df4['YoY, in_percents']=round(df4['Difference']/df4['AveragePrice']*100+100,2)
df4.head()
```

## Пример 2. Расчет ежемесячных темпов роста

```python
df5=df2.copy() # Создается новый data frame
df5
```

```python
df5['Date']=pd.to_datetime(df5['Date']) # Тип данных в колонке Date меняется с текстового на дату и время
df5.head()
```

```python
df5['Year']=df5['Date'].dt.year # Добавляется колонка Year
df5.head()
```

```python
df5['Month']=df5['Date'].dt.month # Добавляется колонка Month
df5.head()
```

```python
df5=df5.groupby(['Year','Month'])['AveragePrice'].mean() # Рассчитываются среднемесячные цены на основе среднедневных
df5.head()
```

```python
df5=df5.reset_index() # Добавляется индекс (см. первую колонку)
df5.head()
```

```python
df5['Difference']=df5['AveragePrice'].diff(1) # рассчитывается абсолютное отклонение по колонке AveragePrice
df5.head()
```

```python
# расчет ежемесячных темпов роста, округление до двух знаков в колонке "MoM, in percents"
df5['MoM, in percents']=round(df5['Difference']/df5['AveragePrice']*100+100,2)
df5.head(12)
```

### Скрипт по расчету ежемесячных темпов роста можно было разместить в одной ячейке

```python
df6=df2.copy()
df6['Date']=pd.to_datetime(df6['Date'])
df6['Year']=df6['Date'].dt.year
df6['Month']=df6['Date'].dt.month
df6=df6.groupby(['Year','Month'])['AveragePrice'].mean()
df6=df6.reset_index()
df6['Difference']=df6['AveragePrice'].diff(1)
df6['MoM, in percents']=round(df6['Difference']/df6['AveragePrice']*100+100,2)
df6.head(12)
```
