# Тепловые матрицы (Heat Map)

```python
sns.heatmap(df_1.corr(method ='pearson'),annot=True,annot_kws={'size': 10});
```

```python
# Другой вариант реализации тепловой матрицы коэффициентов парной корреляции Пирсона

pc = df_1[['Glucose','Outcome','Age']].corr(method ='pearson') # Задаются конкретные переменные (фичи, features)

cols = ['Age','Outcome','Glucose'] # Изменение очередности подписей значений на осях

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30}, # Измененение размера значений
                 cmap="BuPu") # Измененение цвета таблицы inferno BuPu
```
