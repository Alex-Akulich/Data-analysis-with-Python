Если интересующая нас совокупность приблизительно нормально распределена, стандартное отклонение дает информацию о доле наблюдений выше или ниже определенных значений. Например, средний рост взрослых мужчин в США составляет 178 см со стандартным отклонением 8 см. Это означает, что большинство мужчин (около 68% при нормальном распределении) имеют рост в пределах  170–186 см) - одно стандартное отклонение - и почти все мужчины (около 95%) имеют рост в пределах 16 см от среднего (162–194 см) - два стандартных отклонения. 

# Ящичковая диаграмма разработана Джоном Тьюки (1915-2000)
![Джон Тьюки](https://upload.wikimedia.org/wikipedia/ru/f/f2/Tukey.jpg)
Именно Джон Тьюки в 1962 г. в своей статье "Будущее анализа данных" призвал к преобразованию статистики. Он предложил новую научную дисциплину под названием "Анализ данных", которая включала статистический вывод в качестве всего лишь одного из компонентов. Тьюки наладил связи с инженерами и вычислительными сообществами, он ввел понятия - разведочный анализ данных, бит, программное обеспечение (sofware)

# Построение модели парной линейной регрессии

```python
X = np.array(df[["alcohol"]])
y = np.array(df[["suicides"]])
model = smf.ols('suicides ~ alcohol', data=df) # для построения линейной регрессии используется библиотека statsmodels
model = model.fit()
model.summary()
```

# Постройте линейную регрессию, используя библиотеку scikit-learn

```python
from sklearn.linear_model import LinearRegression # из библиотеки Scikit-learn импортируется линейную регрессию
```

```python
model = LinearRegression() #создается объект "модель"
```

```python
x=pd.DataFrame (df.alcohol)
y=pd.DataFrame (df.suicides)
```

```python
model.fit(x,y)
```

```python
model.coef_ # Коэффициент регрессии
```

# Постройте линию регрессии

```python
plt.figure(figsize=(10,6))
plt.scatter(df.alcohol,df.suicides)
plt.plot(x, model.predict(x), color = 'red', linewidth=3)
plt.xlabel('Потребление алкоголя, литров')
plt.ylabel('Кол-во суицидов на 100 тысяч жителей')
```
